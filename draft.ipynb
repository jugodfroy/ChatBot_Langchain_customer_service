```python
import sqlite3
con = sqlite3.connect("data/assistance.sqlite3")

#print all data
cur = con.cursor()
cur.execute("SELECT * FROM assistance")
rows = cur.fetchall()
print(rows)

#convert to pandas dataframe
    
import pandas as pd
df = pd.read_sql_query("SELECT * FROM assistance", con)
df.head()
df.iloc[0]['url']
```

```python
display(df.dtypes)
display(df.describe())
display(df.info())
```

```python
df['content_size'] = df.apply(lambda row: len(row['content']), axis=1)
df.head()

```

```python
df['total_size'] = df.apply(lambda row: len(row['url'])+ len(row['title'])+ len(row['content']), axis=1)
df['full_content'] = df.apply(lambda row : 'Titre de la page : "' +row['title'] +'". Contenu de la page : "' + row['content']+'"', axis=1)
df = df.sort_values(by='total_size', ascending=False)
df.iloc[0]['full_content']
```

```python
#Nb of empty content
df2 = df.apply(lambda row: True
               if row['content_size'] == 0 else False, axis=1)

df3 = len(df2[df2 == True].index)
df3
```

```python
#check for duplicates
display(df.duplicated(subset=['url']).sum())
display(df.duplicated(subset=['title']).sum())

#remove duplicates
df = df.drop_duplicates(subset=['title'])
display(df.duplicated(subset=['title']).sum())
```

```python
from langchain.document_loaders import DataFrameLoader
from langchain.vectorstores import Chroma
```

```python
df_loader = DataFrameLoader(df, page_content_column="full_content")
df_document = df_loader.load()
display(df_document)
df_loader
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
embedding_function = SentenceTransformerEmbeddings(
    model_name="all-MiniLM-L12-v2")

#db = Chroma.from_documents(df_document, embedding_function)
```

```python
# save to disk
test = Chroma.from_documents(
    df_document, embedding_function, persist_directory="./data/chromabis")
test
```

```python
#load from disk
db = Chroma(persist_directory="./data/chroma",
             embedding_function=embedding_function)
```

```python
query = "Commentn me désabonné de NETFLIX ?"
matching_docs = db.similarity_search_with_score(query,5)
display(matching_docs)
#if matching_docs[0][1] > 0.2:
#    matching_docs = ["Vide"]
#
#print(type(matching_docs))
```

```python
context_length = 1700
actual_context = 0
urls = []
urls.append(matching_docs[0][0].metadata['url'])
docs = ''

...
```

```python
from langchain.llms import Ollama
from langchain.callbacks.manager import CallbackManager
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate


llm = Ollama(model="mistral:instruct",
             callback_manager=CallbackManager([StreamingStdOutCallbackHandler()]))

prompt = PromptTemplate(
    input_variables=["question", "docs", "urls"],
    template="""
En tant que ChatBot du service client de FREE, ton objectif est de fournir des réponses factuelles, précises et utiles aux questions des clients. Tu dois répondre uniquement en Francais. Utilise les informations extraites des documents du service client pour répondre. Si la réponse à la question n'est pas disponible dans ta base de données, indique clairement que tu ne sais pas, sans inventer de réponse. Après avoir répondu, recommande une ou plusieurs URL pertinentes parmi celles fournies.

Question du client : "{question}"

Base de données : {docs}

URLs utiles : {urls}

Réponse :
""",
)

chain = LLMChain(llm=llm, prompt=prompt)
print("Starting chain")
response = chain.run(
    question="Comment connaitre mon égibilité à la fibre ?", docs=matching_docs[0], urls=urls)
print("Chain finished")
response = response.replace("\n", "")
```
